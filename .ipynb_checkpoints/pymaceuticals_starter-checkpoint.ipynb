## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import os

# Study data files
mouse_metadata = "C:/Users/Brendan/Documents/GitHub/pymecueticals/Mouse_metadata.csv"
study_results = "C:/Users/Brendan/Documents/GitHub/pymecueticals/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
merged_data = pd.merge(mouse_metadata, study_results, how="outer", on="Mouse ID")
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_table = merged_data.groupby(['Drug Regimen'])
mean_table=summary_table.mean()["Tumor Volume (mm3)"]
mean_table=pd.DataFrame(mean_table)
mean_table.rename(columns= {"Tumor Volume (mm3)":"Mean"}, inplace=True) 
```

```python
median_table=summary_table.median()["Tumor Volume (mm3)"]
median_table=pd.DataFrame(median_table)
median_table.rename(columns= {"Tumor Volume (mm3)":"Median"}, inplace=True) 
```

```python
var_table=summary_table.var()["Tumor Volume (mm3)"]
var_table=pd.DataFrame(var_table)
var_table.rename(columns= {"Tumor Volume (mm3)":"Variance"}, inplace=True) 
```

```python
stdev_table=summary_table.std()["Tumor Volume (mm3)"]
stdev_table=pd.DataFrame(stdev_table)
stdev_table.rename(columns= {"Tumor Volume (mm3)":"Standard Deviance"}, inplace=True) 
```

```python
sem_table=summary_table.sem()["Tumor Volume (mm3)"]
sem_table=pd.DataFrame(sem_table)
sem_table.rename(columns= {"Tumor Volume (mm3)":"Standard Error"}, inplace=True) 
```

```python
test=pd.merge(mean_table, median_table, how="outer", on=["Drug Regimen"])
test1=pd.merge(test, var_table, how="outer", on=["Drug Regimen"])
test2=pd.merge(test1, stdev_table, how="outer", on=["Drug Regimen"])
Final_table=pd.merge(test2,sem_table, how="outer", on=["Drug Regimen"])
pd.set_option('display.max_rows'
              ,500)
Final_table
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers.
sorted_data = merged_data.sort_values(["Drug Regimen", "Mouse ID", "Timepoint"], ascending=True)
data_45 = sorted_data[sorted_data["Timepoint"] == 45]
```

```python
cap_data = data_45[data_45['Drug Regimen'].isin(['Capomulin'])]
cap_list = cap_data.sort_values(["Tumor Volume (mm3)"], ascending=True).reset_index()
cap_list = cap_list["Tumor Volume (mm3)"]
quartiles = cap_list.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of temperatures is: {lowerq}")
print(f"The upper quartile of temperatures is: {upperq}")
print(f"The interquartile range of temperatures is: {iqr}")
print(f"The the median of temperatures is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
capomulin_ = merged_data[merged_data["Drug Regimen"] == "Capomulin"]
capomulin_ = capomulin_.reset_index()
capomulin_
```

```python
m601 = capomulin_.loc[capomulin_["Mouse ID"] == "m601"]
m601
m601 = m601[["Timepoint", "Tumor Volume (mm3)"]]
m601 = m601.reset_index(drop=True)
mouse_line = m601.set_index('Timepoint').plot(figsize=(5, 5), linewidth=3, color='blue')
mouse_line.set_ylabel("Tumor Volume (mm3)")
mouse_line
```

```python
reduced_cap = capomulin_.loc[:, ["Mouse ID", "Weight (g)", "Tumor Volume (mm3)"]]
reduced_cap.head()
```

```python
average_cap = pd.DataFrame(reduced_cap.groupby(["Mouse ID", "Weight (g)"])["Tumor Volume (mm3)"].mean()).reset_index()
average_cap.head()
```

```python
average_cap = average_cap.rename(columns={"Tumor Volume (mm3)": "Average Vol"})
average_cap.head()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
average_cap.plot(kind="scatter", x="Weight (g)", y="Average Vol", grid=True, figsize=(4,4),
              title="Weight Vs. Average Tumor Volume")
plt.show()
```

```python
average_cap
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen

mouse_weight = average_cap["Weight (g)"]
avg_tumor_volume = average_cap["Average Vol"]
correlation = st.pearsonr(mouse_weight,avg_tumor_volume)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
from scipy.stats import linregress
x_values = average_cap['Weight (g)']
y_values = average_cap['Average Vol']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5,10),fontsize=15)
plt.xlabel('Mouse Weight')
plt.ylabel('Average Tumor Volume')
plt.title('Average Tumor vs Mouse Weight')
plt.show()
```
